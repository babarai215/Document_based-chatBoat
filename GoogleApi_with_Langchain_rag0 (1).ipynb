```python

import warnings
#from pathlib import Path as p
from langchain import PromptTemplate
from langchain.chains.question_answering import load_qa_chain
from langchain.document_loaders import PyPDFLoader
from langchain.text_splitter import RecursiveCharacterTextSplitter
from langchain.vectorstores import Chroma
from langchain.chains import RetrievalQA
```

```python
from langchain_google_genai import ChatGoogleGenerativeAI
```

```python
model = ChatGoogleGenerativeAI(model= "gemini-pro",google_api_key=GOOGLE_API_KEY,
                               temperature=0.2, convert_system_message_to_human=True)
```

```python
from google.colab import files
upl = files.upload()
```

Extract text from the pdf

```python
pdf_loader = PyPDFLoader("python-crash-course.pdf")
pages = pdf_loader.load_and_split()
print(pages[3].page_content)
```

```python
len(pages)
```

```python
text_splitter = RecursiveCharacterTextSplitter(chunk_size=1000,chunk_overlap=100)
context = "\n\n".join(str(p.page_content) for  p in pages)
texts = text_splitter.split_text(context)
```

RAG Pipline: Embedding+Gemini(LLM)

```python
from langchain_google_genai import GoogleGenerativeAIEmbeddings
```

```python
embedding = GoogleGenerativeAIEmbeddings(model="models/embedding-001",google_api_key=GOOGLE_API_KEY)
```

```python
vector_index = Chroma.from_texts(texts,embedding).as_retriever(search_kwargs={"k":3})
```

```python
qa_chain= RetrievalQA.from_chain_type(model, retriever=vector_index, return_source_documents=True)
```

```python
question = "what is list"
result= qa_chain({"query":question})
result['result']
```

```python
Markdown(result['result'])
```

```python
result['source_documents']
```

```python
template = """Use the following pieces of context to answer the question at the question at the end. if you don't answer, just say thats sorry, don't know.
           {context}
           Question:{question}
           Helpful Answer:"""

QA_CHAIN_PROMPT= PromptTemplate.from_template(template)

qa_chain = RetrievalQA.from_chain_type(
           model,
           retriever = vector_index,
           return_source_documents=True,
           chain_type_kwargs={"prompt":QA_CHAIN_PROMPT})
```

```python
question = "what is list"
result = qa_chain({"query":question})
result['result']
```
